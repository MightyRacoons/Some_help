```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
lambda_1 = 10**-6 #Интенсивность поломки телефонов
lambda_2 = 10**-6 #Интенсивность отказа приложения
lambda_3 = 10**-7 #Интенсивность отказа сервера ФНС
lambda_4 = 10**-6 #Интенсивность отказа сервера БД1
lambda_5 = 10**-6 #Интенсивность отказа сервера БД2
lambda_6 = 10**-7 #Интенсивность отказа сервева Я.Метрика_1
lambda_7 = 10**-7 #Интенсивность отказа сервера Я.Метрика_2
lambda_8 = 10**-10 #Интенсивность отказа ДЦ
```

```python
def calc_prob(l, t):
    p = np.exp(-1*l*t)
    return p
```

```python
time = np.arange(0, 10**10, step=100)
p1 = calc_prob(lambda_1, time)
p2 = calc_prob(lambda_2, time)
p3 = calc_prob(lambda_3, time)
p4 = calc_prob(lambda_4, time)
p5 = calc_prob(lambda_5, time)
p6 = calc_prob(lambda_6, time)
p7 = calc_prob(lambda_7, time)
p8 = calc_prob(lambda_8, time)
full_proba = p1*p2*p8*(1-(1-p3)*(1-p4)*(1-p5)*(1-p6*p7))
```

```python
plt.figure(figsize=(20,10))
plt.plot(time, full_proba)
plt.plot(time, 1 - full_proba)
t_mean = scipy.trapz(full_proba, time)
plt.vlines(t_mean, 0, 1,linestyles='dashed')
plt.xlabel('Время, ч')
plt.ylabel('Вероятность')
plt.legend(
    [
        'P безотказной работы', 
        'P отказа', 
        'Среднее время безотказной работы {t} часов'.format(t=int(t_mean))
    ]
)
plt.title("Характеристики системы")
plt.show()
```

```python
def elems_plot(probs):
    plt.figure(figsize=(40,20))
    for i in range(len(probs)):
        plt.subplot(2, 4, i + 1)
        plt.plot(time, probs[i])
        plt.plot(time, 1 - probs[i])
        t_mean = scipy.trapz(probs[i], time)
        plt.vlines(t_mean, 0, 1,linestyles='dashed')
        plt.xlabel('Время, ч')
        plt.ylabel('Вероятность')
        plt.legend(
            [
                'P безотказной работы', 
                'P отказа', 
                'Среднее время безотказной работы {t} часов'.format(t=int(t_mean))
            ]
        )
        plt.title("Характеристики элемента {i}".format(i=(i+1)))
    plt.show()
```

```python
time = np.arange(0, 1000000, step=100)
p1 = calc_prob(lambda_1, time)
p2 = calc_prob(lambda_2, time)
p3 = calc_prob(lambda_3, time)
p4 = calc_prob(lambda_4, time)
p5 = calc_prob(lambda_5, time)
p6 = calc_prob(lambda_6, time)
p7 = calc_prob(lambda_7, time)
p8 = calc_prob(lambda_8, time)
probs = [p1, p2, p3, p4, p5, p6, p7, p8]
elems_plot(probs)
```

```python
def tmp(l, l2, t):
    p = np.exp(-1*l*t)
    p2 = np.exp(-1*l2*t)
    pair_f = (p*(1-p)*(1-p))*3
    triple_f = (1-p)*(1-p)*(1-p)
    res_p = p2*(1 - pair_f*triple_f)
    return res_p
```

```python
vo_lambda = 10**-90
time = np.arange(0, 1000000, step=100)
p1 = calc_prob(lambda_1, time)
p2 = tmp(lambda_2, vo_lambda, time)
p3 = calc_prob(lambda_3, time)
p4 = tmp(lambda_4, vo_lambda, time)
p5 = calc_prob(lambda_5, time)
p6 = tmp(lambda_6, vo_lambda, time)
p7 = tmp(lambda_7, vo_lambda, time)
p8 = calc_prob(lambda_8, time)
full_proba = p1*p2*p8*(1-(1-p3)*(1-p4)*(1-p5)*(1-p6*p7))
```

```python
plt.figure(figsize=(20,10))
plt.plot(time, full_proba)
plt.plot(time, 1 - full_proba)
t_mean = scipy.trapz(full_proba, time)
plt.vlines(t_mean, 0, 1,linestyles='dashed')
plt.xlabel('Время, ч')
plt.ylabel('Вероятность')
plt.legend(
    [
        'P безотказной работы', 
        'P отказа', 
        'Среднее время безотказной работы {t} часов'.format(t=int(t_mean))
    ]
)
plt.title("Характеристики системы")
plt.show()
```
